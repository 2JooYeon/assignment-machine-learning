# Linear Regression 

## import library

```python
import numpy as np
import matplotlib.image as img
import matplotlib.pyplot as plt
import matplotlib.colors as colors
from mpl_toolkits.mplot3d import Axes3D
```

## load point data 

```python
filename    = 'assignment_06_data.csv'
data_load   = np.loadtxt(filename, delimiter = ',')

x   = data_load[0, :]
y   = data_load[1, :]

plt.figure(figsize=(8,8))

plt.plot(x, y, '.', color = 'blue')
plt.title('data points')

plt.tight_layout()
plt.show()
```

## compute the loss function

```python
def compute_loss(x, y, theta0, theta1):

    n = len(x)
    diff_square_sum = 0
    loss = 0
    
    for i in range(0, n):
        diff_square_sum += (theta0 + theta1*x[i] - y[i])**2

    loss = diff_square_sum / (2*n)
    
    return loss
```

## compute the gradient for each model parameter

```python
def compute_gradient_theta0(x, y, theta0, theta1):

    n = len(x)
    diff_sum = 0
    dL = 0
    
    for i in range(0, n):
        diff_sum += (theta0 + theta1*x[i] - y[i])
    
    dL = diff_sum / n
    
    return dL
```

```python
def compute_gradient_theta1(x, y, theta0, theta1):

    n = len(x)
    diff_sum = 0
    dL = 0
    
    for i in range(0, n):
        diff_sum += (theta0 + theta1*x[i] - y[i])*x[i]
    
    dL = diff_sum /n

    return dL
```

## gradient descent for each model parameter

```python
num_iteration       = 1000
learning_rate       = 0.01

theta0              = 0
theta1              = 0

theta0_iteration    = np.zeros(num_iteration)
theta1_iteration    = np.zeros(num_iteration)
loss_iteration      = np.zeros(num_iteration)

for i in range(num_iteration):

    now_theta0 = theta0 
    now_theta1 = theta1
    
    theta0  = theta0 - learning_rate * compute_gradient_theta0(x, y, now_theta0, now_theta1)
    theta1  = theta1 - learning_rate * compute_gradient_theta1(x, y, now_theta0, now_theta1)
    loss    = compute_loss(x, y, now_theta0, now_theta1)

    theta0_iteration[i] = theta0
    theta1_iteration[i] = theta1
    loss_iteration[i]   = loss

    print("iteration = %4d, loss = %5.5f" % (i, loss))
```

```python
# result for the optimal theta0 and theta1
min_loss = min(loss_iteration)
optimal_index = loss_iteration.argmin()
optimal_theta0 = theta0_iteration[optimal_index]
optimal_theta1 = theta1_iteration[optimal_index]

f = optimal_theta0 + optimal_theta1 * x
```

## plot the results

```python
def plot_data_regression(x, y, f):

    plt.figure(figsize=(8,6))
    plt.title('linear regression result')
    plt.plot(x, y, '.', color = 'blue')
    plt.plot(x, f,'-', color = 'red')
    
    plt.tight_layout()
    plt.show()
```

```python
def plot_loss_curve(loss_iteration):

    plt.figure(figsize=(8,6))
    plt.title('loss curve')

    iteration = np.arange(0, 1000, 1)
    plt.plot(iteration, loss_iteration, '-', color = 'red')
    
    plt.tight_layout()
    plt.show()
```

```python
def plot_model_parameter(theta0_iteration, theta1_iteration):

    plt.figure(figsize=(8,6))
    plt.title('model parameter')

    iteration = np.arange(0, 1000, 1)
    plt.plot(iteration, theta0_iteration, '-', color = 'blue')
    plt.plot(iteration, theta1_iteration, '-', color = 'green')
    
    plt.tight_layout()
    plt.show()
```

```python
X0  = np.arange(-10, 10, 0.1)

X1  = np.arange(-10, 10, 0.1)

grid_theta0, grid_theta1 = np.meshgrid(X0, X1)

grid_loss = compute_loss(x, y, grid_theta0, grid_theta1)


def plot_loss_surface(grid_theta0, grid_theta1, grid_loss):

    fig = plt.figure(figsize=(8,8))
    ax = fig.add_subplot(projection='3d')
    plt.title('loss surface')
    
    ax.plot_surface(grid_theta0, grid_theta1, grid_loss)

    plt.tight_layout()
    plt.show()
```
